````python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
````

````python
df = pd.read_csv("travel-times.csv")
````

````python
# print shape
print(df)
df.shape # this is shape of dataset
````

````python
# print first 15 rows
df.head(15)
````

````python
df.info()
````

````python
# information for the features (columns) with missing values
df.isnull().sum()
````

````python
# drop the duplicates by keeping only last instance
df.drop_duplicates(keep='last', inplace=True)
````

````python
# calculate the total number of missing values (if any) on each column
df.isnull().sum()
````

````python
# create two copies of the dataframe, and then
df1 = df.copy()
df2 = df.copy()
````

````python
# drop rows with missing values from the 1st copy
df1.dropna()
````

````python
# drop columns with missing values from the 2nd copy
df2.dropna(axis=1)
````

````python
# get summary statistics and see the correlation between the numerical columns
df.describe()
````

````python
# show rows 11 to 14
print(df.iloc[11:15])
````

````python
# create a subset with trips occurred on November 23, 2011 and January 6, 2012
print(df.columns) # to see the name of the column
print()
subset = df[df['Date'].isin(['11/23/2011', '1/6/2012'])]
print(subset)
````

````python
# produce a scatterplot between Distance and TotalTime
df.plot(kind="scatter",x="Distance",y="TotalTime", title="Distance vs Total Time")
````
